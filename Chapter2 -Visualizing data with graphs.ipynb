<strong> The axis() function:</strong>  The function returned a tuple with four numbers corresponding to the
range for the x-axis (0.0, 12.0) and the y-axis (53.0, 57.5). These are the

Similarly, you can use xmin, xmax, and ymax to set the minimum and

```python
import matplotlib.pyplot
def create_graph():
    x_numbers = [1, 2, 3]
    y_numbers = [2, 4, 6]
    matplotlib.pyplot.plot(x_numbers, y_numbers)
    matplotlib.pyplot.show()
if __name__=='__main__':
    create_graph()
```

```python
import matplotlib.pyplot as plt
# Draw the graph
def draw_graph(x, y):
    plt.plot(x, y, marker='o')
    plt.xlabel('Distance in meters')
    plt.ylabel('Gravitational force in newtons')
    plt.title('Gravitational force and distance')
    plt.show()
def generate_F_r():
    # Generate values for r
    r = range(100, 1001, 50)
    # Empty list to store the calculated values of F
    F = []
    # Constant, G
    G = 6.674*(10**-11)
    # Two masses
    m1 = 0.5
    m2 = 1.5
    # Calculate force and add it to the list, F
    for dist in r:
        force = G*(m1*m2)/(dist**2)
        F.append(force)
    # Call the draw_graph function
    draw_graph(r, F)
if __name__=='__main__':
    generate_F_r()
```
